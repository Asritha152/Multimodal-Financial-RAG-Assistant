```python
from groq import Groq
import base64

client = Groq(api_key=groqapi)

def encode_image(image_path):
    with open(image_path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf-8")

for ele in elements:
    if isinstance(ele, Image):

        image_path = ele.metadata.image_path
        base64_image = encode_image(image_path)

        response = client.chat.completions.create(
            model="meta-llama/llama-4-scout-17b-16e-instruct",
            messages=[
                {
                    "role": "user",
                    "content": [
                        {
                            "type": "text",
                            "text": """
                            You are generating descriptions for a financial RAG system.

                            Analyze this image thoroughly and extract:

                            1. Visual type (bar chart, line chart, pie chart, infographic, table, diagram, etc.)
                            2. Title and subtitles
                            3. All visible labels and legends
                            4. Numerical values, percentages, and dates
                            5. Trends, comparisons, and relationships
                            6. Key business or financial insights
                            7. A concise summary

                            Be factual and comprehensive.
                            """
                        },
                        {
                            "type": "image_url",
                            "image_url": {
                                "url": f"data:image/png;base64,{base64_image}"
                            }
                        }
                    ]
                }
            ]
        )

        image_chunks.append({
            "content": response.choices[0].message.content,
            "metadata": {
                **ele.metadata.to_dict(),
                "element_type": "image",
                "source_type": "vision_summary"
            }
        })
```

```python
image_chunks
```

```python
text_chunks = []

for ele in elements:
    if isinstance(ele, (NarrativeText, Title, ListItem)):

        text_chunks.append({
            "content": ele.text,
            "metadata": {
                **ele.metadata.to_dict(),
                "element_type": type(ele).__name__
            }
        })
```

```python
table_chunks=[]
```

```python
TABLE_PROMPT='''You are a financial analyst creating a retrieval-optimized summary of a t
Use the plain text to understand the column names and the HTML for row st
Plain text of table:
{table_plain}
HTML of table:
{table_html}
Write 4-6 plain sentences (no markdown bullets or headers) that:
1. State clearly what this table is about
2. Name every metric and explicitly state its value for each year
3. Describe which values are highest, lowest, and the year-over-year chan
4. Use natural language so this summary can be found by semantic search
Summary:'''
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser

llm = ChatGroq(model="llama-3.3-70b-versatile",api_key=groqapi)
```

```python
prompt=PromptTemplate(template=TABLE_PROMPT,input_variables=["table_plain","table_html"])
```

```python
parser=StrOutputParser()
```

```python
chain=prompt|llm|parser
```

```python
table_chunks=[]
```

```python
for ele in elements:
    if isinstance(ele, Table):
      response=chain.invoke({"table_plain":ele.text,"table_html":ele.metadata.text_as_html})
      table_chunks.append({
          "content":response,
          "metadata":{
              **ele.metadata.to_dict(),
              "element_type": "table",
        "source_type": "table_summary"
          }})
```

```python
table_chunks
```

```python
chunks=text_chunks+table_chunks+image_chunks
```

```python
len(chunks)
```

```python
docs=[]
```

```python
from langchain_core.documents import Document
```

```python
for chunk in chunks:
  docs.append(Document(page_content=chunk["content"],metadata=chunk["metadata"]))
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-2-preview",api_key=userdata.get('GOOGLE_API_KEY'))
```

```python
!pip install langchain_chroma
```

```python
from langchain_chroma import Chroma
from langchain_community.vectorstores.utils import filter_complex_metadata

docs = filter_complex_metadata(docs)
vectorstore = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    persist_directory="./chroma_db"
)
```

```python
retreiver=vectorstore.as_retriever(
    search_kwargs={"k": 5}
)
```

```python
retrieved_docs=retreiver.invoke('What are the key risks mentioned in the FY2025 outlook?')
```

```python
retrieved_docs
```

```python
context_parts = []

for doc in retrieved_docs:

    content = doc.page_content

    if doc.metadata.get("text_as_html"):
        content += (
            "\n\nTable HTML:\n"
            + doc.metadata["text_as_html"]
        )

    context_parts.append(content)

context = "\n\n".join(context_parts)
```

```python
context
```

```python
ragprompt=PromptTemplate(
    template="""
    You are a financial document assistant.
Answer the user's question using ONLY the information provided in the retrieved context.

Retrieved Context:
{context}

User Question:
{question}

Instructions:

* Use information from all available sources, including text sections, table summaries, figure captions, and image descriptions.
* If numerical values are present, include them accurately.
* If multiple retrieved chunks provide complementary information, combine them into a single coherent answer.
* Do not make up facts that are not present in the context.
* If the answer cannot be determined from the context, say "The provided documents do not contain enough information to answer this question."
* When relevant, mention the source type (table, figure, or text section) naturally in the explanation.
* Keep the answer concise but complete.

Answer:

    """
,input_variables=['context','question'])
```

```python
rag_chain=ragprompt|llm|parser
```

```python
ans=rag_chain.invoke({
    "context":context,
    "question":'What are the key risks mentioned in the FY2025 outlook?'

})
```
